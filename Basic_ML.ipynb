#Numpy:
NumPy is a tool in Python that helps us work with large groups of numbers more easily and quickly.

In simple terms:

- **Big Arrays**: It lets us create big lists of numbers (arrays) that are faster to work with than regular Python lists.
- **Math Functions**: It comes with a bunch of math tools so we can add, multiply, or do other calculations with whole arrays at once.
- **Works Well with Other Tools**: It’s like a building block that other tools (like Pandas for data tables) are built on.

```python
import numpy as np
```

```python
# Create a 1D array
array = np.array([1, 2, 3, 4, 5])
print("Array:", array)
```

```python
#With NumPy, we can perform operations on the entire array at once.
# Add 10 to each element
array_plus_ten = array + 10
print("Array + 10:", array_plus_ten)

# Multiply each element by 2
array_times_two = array * 2
print("Array * 2:", array_times_two)
```

```python
#We can create a 2D array (like a matrix) and perform operations on it.
# Create a 2D array
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\\n", matrix)

# Transpose the matrix (switch rows and columns)
transposed = matrix.T
print("Transposed Matrix:\\n", transposed)
```

```python
#NumPy also has useful functions like finding the sum, mean, and maximum.

# Sum of all elements in the array
sum_array = np.sum(array)
print("Sum:", sum_array)

# Mean (average) of the array
mean_array = np.mean(array)
print("Mean:", mean_array)

# Maximum element in the array
max_element = np.max(array)
print("Max element:", max_element)
```

#Pandas
**Pandas** is a Python library for working with and **analyzing data in a structured, tabular format**. It allows us to easily **manipulate data, perform calculations, and analyze data** with powerful tools. The main structure in Pandas is called a **DataFrame**, which is like a table with **rows and columns**, similar to a **spreadsheet**.

```python
import pandas as pd
```

```python
#Create a DataFrame with some sample data, like a small table of student grades.

# Sample data as a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Math": [85, 78, 90, 92],
    "Science": [88, 92, 79, 85],
    "English": [95, 80, 85, 90]
}

# Creating a DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# Calculate the mean for each subject
average_grades = df.mean(numeric_only=True)
print("Average Grades:\\n", average_grades)
```

```python
#Filter to find students who scored above 90 in Math.
high_math_scores = df[df["Math"] > 90]
print("Students with high Math scores:\\n", high_math_scores)
```

```python
#Add a column for the Average score of each student.
# Calculate the average score for each student and add it as a new column
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)
print("Data with Average Scores:\\n", df)
```

#Matplotlib
**Matplotlib** is a Python library used for creating **visualizations, like plots and charts**. It’s particularly popular for making line plots, bar charts, scatter plots, histograms, and more, and is often used in data analysis to help visualize data.

```python
import matplotlib.pyplot as plt
```

```python
#We have some data about the population growth of a city over five years. We’ll use Matplotlib to plot this data.

# Years and population data
years = [2018, 2019, 2020, 2021, 2022]
population = [1.5, 1.7, 1.9, 2.1, 2.3]

# Create a line plot
plt.plot(years, population)

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.title("City Population Growth Over Time")

# Show the plot
plt.show()
```

```python
#We can make a bar chart for the same data to illustrate how Matplotlib works with different chart types.


# Create a bar chart
plt.bar(years, population, color='skyblue')

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.title("City Population Growth Over Time (Bar Chart)")

# Show the plot
plt.show()

```

```python
#We can customize your charts with colors, line styles, markers, legends, and more. Here’s an example with markers:

# Line plot with markers
plt.plot(years, population, marker='o', linestyle='-', color='purple')

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.title("City Population Growth Over Time")

# Show the plot
plt.show()
```
